```python
eps_0 = 2.37
F_0 = 0.14
ess = 0.37

def epsilon_g(F_planet):
    return eps_0 * np.exp(-1* (np.log10(F_planet) - F_0)**2 / (2*ess**2))
```

```python
import jaxopt
from tinygp import GaussianProcess, kernels, transforms


def train_gp(nparams, build_gp_func):
    @jax.jit
    def loss(params):
        return -build_gp_func(params).log_probability(y)

    params = {
        "log_amp": np.float64(0.0),
        "log_scale": np.zeros(nparams),
    }
    solver = jaxopt.ScipyMinimize(fun=loss)
    soln = solver.run(params)
    return build_gp_func(soln.params)
```

```python
yerr=100.0*np.ones_like(y)
```

```python
def build_gp_corr(params):
    kernel = jnp.exp(params["log_amp"]) * transforms.Cholesky.from_parameters(
        jnp.exp(params["log_scale"][:2]),
        params["log_scale"][2:],
        kernels.ExpSquared(),
    )
    return GaussianProcess(kernel, X, diag=yerr**2)


corr_gp = train_gp(3, build_gp_corr)
```

```python
y_pred = corr_gp.condition(y, X_pred).gp.loc.reshape((100, 200))
```

```python
flux_pred = (4 * const.sigma_sb * (y_pred*u.K) **4).to(u.Gerg/u.s/u.cm**2).value
```

### Add a second panel to compare to XUV photoionization

We want to mimic Thorngren et al. 2023 figure!

```python
R_star = 2.546 * u.Rsun
M_star = 1.642 * u.Msun
a_by_R_star = 5.691 
a_AU = (a_by_R_star * R_star).to(u.AU)
m_planet = (0.34 * u.Mjup).to(u.Msun)
```

```python
r_Hill = a_AU * (m_planet/(3*M_star))**(1/3)
```

```python
R_planet = 2.085 * u.R_jup
```

```python
rH_by_Rp = (r_Hill/R_planet).to(u.dimensionless_unscaled)
rH_by_a = (r_Hill/a_AU).to(u.dimensionless_unscaled)
```

```python
xi = rH_by_Rp
```

```python
Kt = 1 - 3/(2*xi) + 1/(2*xi**3)
```

```python
eta = 0.4 # "~tens of percent"
```

```python
Fxuv =  10**-4.16 * (3e9 * u.erg / u.s / u.cm**2)
```

```python
rho_planet = m_planet / (4/3 * np.pi * R_planet**3)
rho_planet = rho_planet.to(u.g/u.cm**3)
```

```python
M_dot = 0.75 * eta * (Fxuv) / (const.G * Kt * rho_planet)
```

Compute density

```python
density_map = 10**x_*u.Mjup/(4/3*np.pi*(y_ * u.R_jup)**3)
```

```python
M_dot_map = 0.75 * eta * (Fxuv) / (const.G * Kt * density_map)
```

```python
ionization_timescale = 10**x_*u.Mjup / M_dot_map
```
