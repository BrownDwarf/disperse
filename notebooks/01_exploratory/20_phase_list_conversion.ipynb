```python
df = pd.read_csv('C:/Users/jojoa/GitHub/disperse/data/HPF/phase_tracks/east_track.csv')
```

```python
df['btjd_east_start']=df['Start JD']-2457000.0 #cut down JD to smaller scale
```

```python
target_name = 'hat-p-67b'
orbital_period = 4.81010250
t_c = 2455961.38467
t_c_btjd = t_c-2457000.0 #neg value means discovered before TESS
```

```python
df['phase_east_start']=np.mod(df.btjd_east_start-t_c_btjd+(orbital_period*0.5),orbital_period)-0.5*orbital_period
df['normalized_phase_east_start']=(df['phase_east_start']/orbital_period) #centered at 0
```

```python
df['btjd_east_end']=df['End JD']-2457000.0 #cut down JD to smaller scale
```

```python
df['phase_east_end']=np.mod(df.btjd_east_end-t_c_btjd+(orbital_period*0.5),orbital_period)-0.5*orbital_period
df['normalized_phase_east_end']=(df['phase_east_end']/orbital_period) #centered at 0
```
